```python
documents = {
    "S1": "The best homemade pizza recipe for beginners",
    "S2": "How to make the best pizza dough from scratch recipe",
    "S3": "Top 10 tips for baking pizza at home for beginners",
    "S4": "Easy pasta recipes and Italian cooking techniques",
    "S5": "A guide to building your own outdoor pizza oven in summer",
    "S6": "The history of pizza from Naples to New York for foodies",
    "S7": "Healthy meal prep ideas and cooking tips for temperature control",
    "S8": "Best wood-fired pizza restaurants in the best cities",
    "S9": "How to choose the right flour for homemade bread and pizza",
    "S10": "Italian cheese varieties and their uses in traditional cooking"
}
```

```python
from google.colab import drive
drive.mount('/content/drive')

#PARSING THE HTML TEXT
from bs4 import BeautifulSoup
import pandas as pd

with open('/content/drive/MyDrive/Colab Notebooks/index.html', 'r') as f:
    contents = f.read()

    soup = BeautifulSoup(contents, "html.parser")
    results = soup.find_all('li')

    df = pd.DataFrame({'Names': results})
    df.to_csv('names.csv', index=False, encoding='utf-8')
```

```python
text = soup.get_text(separator='\n', strip=True)
print(text)
```

```python
# Convert text from index.html to dictionary in python and save it under documents dictionary
# Splits by comma first, then by colon for each pair
documents = {k.strip(): v.strip() for k, v in (item.split(':') for item in text.split(','))}

print (documents)
```

# **STOP WORDS**

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
#print(stopwords.words('english'))
```

```python
len(stopwords.words('english'))
```

```python
STOPWORDS = stopwords.words('english')

for doc_id, text in documents.items():
  tokenized = text.lower().split()
  print("Before : ", tokenized)
  print("After : ",[word for word in tokenized if word not in STOPWORDS])
```

## **STEMMING & LEMMATIZATION**

```python
from nltk.stem import PorterStemmer
```

```python
%%time
PorterStemmer().stem('reading')
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
%%time
WordNetLemmatizer().lemmatize('reading')
```

```python
%%time
for doc_id, text in documents.items():
  tokenized = text.lower().split()
  print("Before : ", tokenized)
  print("After : ",[PorterStemmer().stem(word) for word in tokenized if word not in STOPWORDS])
```

```python
%%time
for doc_id, text in documents.items():
  tokenized = text.lower().split()
  print("Before : ", tokenized)
  print("After : ",[WordNetLemmatizer().lemmatize(word) for word in tokenized if word not in STOPWORDS])
```

```python
processed = {}
for doc_id, text in documents.items():
  tokenized = text.lower().split()
  print("Before : ", tokenized)
  print("After : ",[WordNetLemmatizer().lemmatize(word) for word in tokenized if word not in STOPWORDS])
  processed[doc_id] = [WordNetLemmatizer().lemmatize(word) for word in tokenized if word not in STOPWORDS]
```

```python
processed
```

```python
unique_words = set()
for doc_id, text in processed.items():
   unique_words.update(text)
print(unique_words)
print(len(unique_words))
```

```python
from collections import defaultdict
inverted_index = defaultdict(set)

for doc_id, text in processed.items():
  for word in text:
    inverted_index[word].add(doc_id)
print (inverted_index)
```

```python
for word in unique_words:
  docs = sorted(inverted_index[word])
  print(word, " : ", docs)
```
